```python
testing_lbwo = pd.read_excel('C://Users//S1130863//Downloads/diss/testing_lbwo.xlsx')
```

```python
testing_last2years = testing_lbwo[(testing_lbwo['Observation Year']>=2000)]
```

```python
y_real = testing_last2years.loc[:,['LB_binary']]
y_apriori = testing_last2years.loc[:,['Apriori-rule']]
```

```python
y_real
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_real, y_apriori)
print(cm)
accuracy_score(y_real, y_apriori)
```

```python
from sklearn.metrics import recall_score, precision_score
recall_score(y_real, y_apriori)
```

```python
precision_score(y_real, y_apriori)
```

```python
from sklearn.metrics import f1_score
f1_score(y_real,y_apriori)
```

### Just with apriori rules we are getting 92% recall and 45% prec and 61% f1.
