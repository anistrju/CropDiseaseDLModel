```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_excel('C://Users//S1130863//Downloads/diss/curated_data.xlsx',sheet_name = 'LB_NoOthers_Final')
```

```python
dataset
```

```python
years = dataset.loc[:,'Observation Year'].unique().tolist()
```

```python
import random
selectedyears = random.sample(years,5)
selectedyears
```

```python
data_train = dataset[dataset['Observation Year'].isin(selectedyears)]
```

```python
data_test = dataset[~dataset['Observation Year'].isin(selectedyears)]
```

```python
data_test
```

```python
X = dataset.loc[:, ['MAXT','MINT','RH1','RH2','RF','WS','SSH','EVP','BPH','GM','GLH','YSB']]
y = dataset.loc[:, ['LB-binary']]
```

```python
X_train = data_train.loc[:, ['MAXT','MINT','RH1','RH2','RF','WS','SSH','EVP','BPH-binary','GM-binary','GLH-binary','YSB-binary']]
X_test = data_test.loc[:, ['MAXT','MINT','RH1','RH2','RF','WS','SSH','EVP','BPH-binary','GM-binary','GLH-binary','YSB-binary']]
y_train = data_train.loc[:, ['LB-binary']]
y_test = data_test.loc[:, ['LB-binary']]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 1)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
#X_train.iloc[:,:8] = sc.fit_transform(X_train.iloc[:,:8])
#X_test.iloc[:,:8] = sc.transform(X_test.iloc[:,:8])
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import recall_score, precision_score
recall_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```

## Results of Logistic Regression

### achieving 58% f1 and 87% accuracy with a purely weather based model and random splitting in 70:30 ratio.

### achieving greater than 91% accuracy and 72% with a weather plus pest data and random splitting in 70:30 ratio.
### Most importantly, we get disastrous results if we pass weather and pest data without pest discretization, 47% f1 score

### achieving greater than 91% accuracy and 74% f1 with year wise splitting in 5:2 ratio and weather and pest data
