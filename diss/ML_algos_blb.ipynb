```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_excel('C://Users//S1130863//Downloads/diss/curated_data.xlsx',sheet_name = 'BLB_final')
```

```python
dataset
```

```python
years = dataset.loc[:,'Observation Year'].unique().tolist()
```

```python
import random
selectedyears = random.sample(years,3)
selectedyears
```

```python
data_train = dataset[dataset['Observation Year'].isin(selectedyears)]
```

```python
data_test = dataset[~dataset['Observation Year'].isin(selectedyears)]
```

```python
data_test
```

```python
X = dataset.loc[:, ['MaxT','MinT','RH1','RH2','RF','WS','SSH','EVP','Th-Binary','Wht-Binary','Jass-Binary','Others-Binary']]
y = dataset.loc[:, ['BLB-Binary']]
```

```python
X_train = data_train.loc[:, ['MaxT','MinT','RH1','RH2','RF','WS','SSH','EVP','Th-Binary','Wht-Binary','Jass-Binary','Others-Binary']]
X_test = data_test.loc[:, ['MaxT','MinT','RH1','RH2','RF','WS','SSH','EVP','Th-Binary','Wht-Binary','Jass-Binary','Others-Binary']]
y_train = data_train.loc[:, ['BLB-Binary']]
y_test = data_test.loc[:, ['BLB-Binary']]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train.iloc[:,:8] = sc.fit_transform(X_train.iloc[:,:8])
X_test.iloc[:,:8] = sc.transform(X_test.iloc[:,:8])
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import recall_score, precision_score
recall_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```

## Results of Logistic Regression

### achieving 90% accuracy with a purely weather based model and random splitting in 75:25 ratio.

### achieving greater than 94% accuracy with a weather plus pest data and random splitting in 75:25 ratio.

### achieving greater than 92% accuracy with year wise splitting in 75:25 ratio and weather and pest data

## KNN with year wise and full data 84% accuracy (recheck)
